# Trying new values & Splitting Data

```python
Diamonds_nums=(Diamonds.select_dtypes(exclude="category").columns).drop('price',1)
Diamonds_cat=Diamonds.select_dtypes(include="category").columns
Diamonds_nums
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
D_ind , T_ind , C_ind = 4, 5,0

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def fit(self,X ,y=None):
        return self
    def transform(self, X):
        D_over_T=X[:,D_ind]/X[:,T_ind]
        C_over_D=X[:,C_ind]/X[:,D_ind]
        return np.c_[X,D_over_T,C_over_D]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler

num_pipeline= Pipeline([
     ("std_scalor", StandardScaler()),
     ('attribs_adder', CombinedAttributesAdder())
])

full_pipeline= ColumnTransformer([
    ("num", num_pipeline, Diamonds_nums),
    ("cat", OrdinalEncoder(), Diamonds_cat)
])
```

```python
y=Diamonds["price"]
x=Diamonds.drop('price', axis=1)
```

```python
x_prepared=full_pipeline.fit_transform(x)
x_prepared=pd.DataFrame(x_prepared,columns= list(x.columns)+["D_over_T", "C_over_D]"],index=x.index)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test=train_test_split(x_prepared,y, test_size=0.2, random_state=42)
```

# Training model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
lin_prediction=lin_reg.predict(x_test)
lin_rmse=np.sqrt(mean_squared_error(y_test, lin_prediction))
lin_acc=lin_reg.score(x_test,y_test)*100
print("Linear Regression")
print("Rmse:  ",lin_rmse)
print("accuracy:  ",lin_acc)
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg=DecisionTreeRegressor()
tree_reg.fit(x_train, y_train)
tree_prediction=tree_reg.predict(x_test)
tree_rmse=np.sqrt(mean_squared_error(y_test, tree_prediction))
tree_acc=tree_reg.score(x_test,y_test)*100
print("Decision Tree Regressor ")
print("Rmse:  ",tree_rmse)
print("accuracy:  ",tree_acc)
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor()
forest_reg.fit(x_train, y_train)
forest_prediction=forest_reg.predict(x_test)
forest_rmse=np.sqrt(mean_squared_error(y_test, forest_prediction))
forest_acc=forest_reg.score(x_test,y_test)*100
print("Random Forest Regrssor")
print("Rmse:  ",forest_rmse)
print("accuracy:  ",forest_acc)
```

# Cross Validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(lin_reg, x_train, y_train, scoring="neg_mean_squared_error", cv=10)
linear_rmse_scores = np.sqrt(-scores)
linear_rmse_scores.mean()
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree_reg, x_train, y_train, scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
tree_rmse_scores.mean()
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(forest_reg, x_train, y_train, scoring="neg_mean_squared_error", cv=10)
forest_rmse_scores = np.sqrt(-scores)
forest_rmse_scores.mean()
```

#### As we saw, random forest model was found to be the best model among the models we tested, with the lowest value of error.
